```python
#Import the necessary modules
from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.wcs import WCS
import warnings
import numpy as np
warnings.filterwarnings('ignore')
from astroquery.vizier import Vizier
from astroquery.gaia import Gaia
from astroquery.simbad import Simbad
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Image
from IPython.core.display import HTML
import pandas as pd
```

```python
c=SkyCoord("11 02 24.8763629208 -77 33 35.667131796", unit=(u.hourangle, u.deg))
c
```

```python
#Gaia.ROW_LIMIT = -1
Vizier.ROW_LIMIT=-1
```

```python
result = Vizier.query_region(SkyCoord(ra=c.ra.deg, dec=c.dec.deg,unit=(u.degree, u.degree), frame='icrs'),
                            radius=u.Quantity(10,u.deg),
                             catalog=["I/355/gaiadr3"]
                            )
result #"B/wds",
```

```python
print(result[0])
```

```python
result[0]["Gmag"]
```

```python
cond_gaia=(result[0]["RUWE"]<1.2)&(result[0]["Gmag"]<7)

print(result[0][cond_gaia])
```

```python
Simbad
```

```python
gaia_id=result[0]["Source"][cond_gaia]
gaia_id
```

```python
all_ids=[]
for item in gaia_id:
    result_table = Simbad.query_objectids("Gaia DR3 %s"%item)#5201179389431267584

    #print(result_table)
    if result_table == None:
        print("skip")
    else:
        all_ids.append(result_table)
```

```python
all_ids
```

```python
#find if the id has WDS 
#element.split
all_nonempty_ids=[]
for item in all_ids:
    if item == None:
        print(item)
    else:
        all_nonempty_ids.append(item)
```

```python
len(all_nonempty_ids)
```

```python
for i in range(len(all_nonempty_ids)):
    for item in all_nonempty_ids[i]:
        #print(item)
        if "WDS" in str(item):
            print("Star with this ID is crossmatched with Gaia and WDS")
            print(all_nonempty_ids[i])
        else:
            print("no WDS found")
```
